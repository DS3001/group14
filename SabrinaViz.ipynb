```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/df_cleaned.csv')
```

```python
df.head()
```

Income and Religion

```python
#Adding a new column with the income range for easier interpretation

income_mapping = {
    5: '$5,000 to $5,999',
    6: '$6,000 to $6,999',
    7: '$7,000 to $7,999',
    8: '$8,000 to $9,999',
    9: '$10,000 to $12,499',
    10: '$12,500 to $14,999',
    11: '$15,000 to $17,499',
    12: '$17,500 to $19,999',
    13: '$20,000 to $22,499',
    14: '$22,500 to $24,999',
    15: '$25,000 to $29,999',
    16: '$30,000 to $34,999',
    17: '$35,000 to $39,999',
    18: '$40,000 to $49,999',
    19: '$50,000 to $59,999',
    20: '$60,000 to $74,999',
    21: '$75,000 to $89,999',
    22: '$90,000 to $109,999',
    23: '$110,000 to $129,999',
    24: '$130,000 to $149,999',
    25: '$150,000 to $169,999',
    26: '$170,000 or Over'
}

df['INCOME_RANGE'] = df['INCOME16'].map(income_mapping)
```

```python
sns.countplot(x='RELIG', data=df)
plt.xlabel('Religion')
plt.ylabel('Count')
plt.title('Distribution of Religion')
plt.xticks(rotation=90)
plt.show()
```

Most people are Protestant in our dataset

```python
sns.boxplot(x='RELIG', y='INCOME16', data=df)
plt.xlabel('Religion')
plt.ylabel('Income')
plt.title('Income Distribution by Religion')
plt.xticks(rotation=90)
plt.show()
```
